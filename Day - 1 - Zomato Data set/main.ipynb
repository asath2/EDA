```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index()
```

```python
# Changeing name 0 to Rating Rate
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
# Observation
    # 1. When rating is between 4.5 to 4.9 -> Excellent
    # 2. When rating is between 4.0 to 4.4 -> Very Good
    # 3. When rating is between 3.5 to 3.9 -> Good
    # 4. When rating is between 2.5 to 3.4 -> Average
    # 5. when rating is between 0.8 to 2.4 -> Poor
    # 6. When rating 0 -> Not rated
    # 0 -> Rating provided by maximum people.
```

```python
ratings.head()
```

```python
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings)
```

```python
# change size

matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings,color='red')
```

```python
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings,hue='Rating')
```

```python
# Color Maping
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','darkgreen'])
```

```python
# Observation:
    # Most  of the peoples are rated 3.0 to 3.5
    # Not rated people count is very high

np.average(ratings['Aggregate rating'])


    # The average rating -> 3.24
```

```python
# Count plot

sns.countplot(x='Rating color',data=ratings,palette=['pink','red','orange','yellow','green','darkgreen'])
```
